# 项目背景  
UserBehavior是阿里巴巴提供的一个淘宝用户行为数据集，数据集包含了2017年11月25日至2017年12月3日之间，有行为的约一百万随机用户的所有行为（行为包括点击、购买、加购、喜欢），用以进行隐式反馈推荐问题的研究。  
注：隐式反馈推荐问题
通过对2017年11月25日至2017年12月3日之间的用户行为数据分析，为客户提供更精准的隐式反馈推荐。  

# 项目目标

## 分析目的

**1.从用户角度：提高用户忠诚度，帮助用户快速找到商品，减轻信息过载带来的问题。**   
**2.从网站角度：提高网站交叉销售能力，提高成交转化率， 从而提高成交额。**  
**3.最终目的：能使得推荐系统能将更多的物品推荐给用户， 能给用户推荐他们真正感兴趣的物品。**  
注：数据来源为下面所示的天池数据集，但是由于数据量过大（约有1亿条用户行为），所以抽样约**10000000**的用户行为进行分析。 

参考资料  
> https://zhuanlan.zhihu.com/p/105766780 《数据分析项目实战——淘宝用户行为分析》-------作者：（知乎）何书365

## 分析思路  
主要进行以下四个专题分析：
1. <font color=red>描述分析</font>:从整体维度和日期维度查看观察窗口内的UV PV PV/UV 成交量， 从用户维度查看用户的行为数据。
2. <font color=red>用户流量行为分析</font>:分析网站用户的各行为转化率。  
3. <font color=red>用户类目偏好分析</font>:从商品的点击、收藏、加购、购买的频率，探索用户对商品的购买偏好，找到针对不同商品的营销策略(什么产品以及产品类目的购买率最高，找出最受欢迎的产品，优化产品推荐)。   
4. <font color=red>用户行为习惯分析</font>:从时间维度分析观察窗口内平均而言一天当中的哪段时间用户数最多， 找到一天当中用户最活跃的时刻。

## 时间戳处理  
在计算机中，时间实际上是用数字表示的。  
我们把1970年1月1日 00:00:00 UTC+00:00时区的时刻称为epoch time(纪元)，记为0（1970年以前的时间timestamp为负数），当前时间就是相对于epoch time的秒数，称为timestamp。   
timestamp的值与时区毫无关系，因为timestamp一旦确定，其UTC时间就确定了，转换到任意时区的时间也是完全确定的，这就是为什么计算机存储的当前时间是以timestamp表示的，因为全球各地的计算机在任意时刻的timestamp都是完全相同的（假定时间已校准）

### 时间维度--按日期维度查看UV、PV、人均浏览次数、成交量

1. UV = Unique Visitor独立访客, 每个日期下的不同用户ID计数
2. PV = page view页面浏览量, 每个日期下pv的计数
3. 人均浏览次数 = PV/UV
4. 成交量 = 每个日期buy的求和

```python
def data_UV_PV_UVPV_cj(DF):
    # 独立访客
    UV = DF['用户ID'].value_counts().size
    # 页面浏览量
    PV = DF['行为类型'].value_counts()['pv']
    # 人均浏览次数
    人均浏览次数 = float(format(PV/UV, '.2f'))
    # 成交量
    成交量 = DF['行为类型'].value_counts()['buy']
    return pd.Series({'UV':UV, 
                         'PV':PV, 
                         '人均浏览次数':人均浏览次数, 
                         '成交量':成交量
                        })
a = df.groupby('日期').apply(data_UV_PV_UVPV_cj)
```

#### 用户跳失率  
跳失率的定义， 用户更具链接点击进入淘宝网页， 我们并不知道统计规则里面这个进来的页面是否会算作pv  
安全起见，我们认为这个算作pv， 但不是主动浏览的， 可能是广告渠道等等。  
所以， 我们定义， **只有pv>=2的客户才不是跳失客户**。但是很奇怪的是有些用户pv没有， 但是却有其他行为， 甚至是购买行为， 这些客户就不能算是跳失客户， 所以， 还同时要求其他行为均为0.

```python
def tree_plot(be):
    a = df[df.行为类型==be]['商品类目ID'].value_counts()[0:20]
    #print(a)
    b = a.to_frame().reset_index()
    b['name'] = b.pop('index')
    b = b.rename(columns = {"商品类目ID":'value'})
    b = b.T.to_dict()
    b = b.values()
    b = list(b)
    # b
    from pyecharts import options as opts
    from pyecharts.charts import TreeMap
    from pyecharts.globals import ThemeType
    data =b
    c = (
        TreeMap(init_opts=opts.InitOpts(theme=ThemeType.LIGHT, width='800px', height='400px'))
        .add("商品类目", data, )
        .set_global_opts(title_opts=opts.TitleOpts(title="{}-top20商品类目".format(be)))
    )
    return c.render_notebook()
```

# 结论

**一**、通过对用户流量行为的转化分析，可以看出用户从点击到购买的转化率还是比较高的，目前来看可以通过引导用户收藏并加购来提高用户从收藏/加购到购买的转化率。而流量行为从点击到购买的转化率仅有2.24%，故从点击到购买的行为转化是一个提高的重点。针对这一环节的建议有：  
- 1 优化电商平台的搜索匹配度和推荐策略，提高筛选精确度，并对搜索和筛选的结果排序的优先级进行优化；  
- 2 可以给客户提供同类产品比较的功能，让用户不需要多次返回搜索结果进行点击查看，方便用户确定心仪产品，增加点击到后续行为的转化；  
- 3 优化收藏到购买的操作过程，增加用户收藏并加购的频率，以提高购买转化率。  

**二**、通过用户行为习惯分析
- 1 可以看出用户的活跃时间高峰期主要在13:00-14:00点，此时使用人数最多，活动最容易触达用户，所以可以将营销活动安排在这个时间段内，来进行引流并转化； 
  
**三**、通过对用户的商品偏好分析，可以看出商品销量主要是依靠长尾效应，而非爆款商品的带动。但是通过对商品品类的分析可以看出能吸引用户注意力的商品购买转化率并不高，是一个提高销量的突破口。针对用户关注度高但销量不高的这部分产品，可以从以下几个方面着手：
- 1 商家在详情页上的展示突出用户重点关注的信息，优化信息的呈现方式，减少用户的时间成本；  
- 2 增加这些产品的质量管控力度，加强对用户反馈建议如评论的管理，认真采纳并根据自身的优劣势进行商品优化  
- 3 此外，对于购买top20的商品，可以在电商首页对这些品类的商品优先进行展现，以满足用户的购买需求。
