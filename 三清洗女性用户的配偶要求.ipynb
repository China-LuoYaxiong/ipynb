# 拆分 择偶条件

## 新增 年龄要求  
位置在子女情况前面

```python
a = df.择偶条件.str.split('|')
a
```

```python
df.insert(19, '年龄要求', '不明')
```

```python
def x(s):
    for i in s:
        if '岁' in i:
            return i
```

```python
a.map(x)
```

```python
df['年龄要求'] = a.map(x)
```

```python
df.info()
```

```python
# 提取出 年龄要求有缺失的行
df[df.年龄要求.isna()]    # 那我们以 不明 填写
```

```python
df.年龄要求.fillna('不明', inplace=True)
```

```python
df.info()
```

```python
set(df.年龄要求.value_counts().index)  # 无异常值
```
