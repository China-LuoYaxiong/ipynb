### 两类样本的年龄分布

```python
sns.kdeplot(df.age[df.resp_flag==1],label='1',shade=True)
sns.kdeplot(df.age[df.resp_flag==0],label='0',shade=True)
plt.xlim([60,90])
plt.xlabel('age')
plt.ylabel('Density');
```

显然年龄在65-70之间买的人明显更多

# 缺失值处理

## 首先统计每一列的数据类型

```python
NA  # 所有缺失的特征，共20个
```

```python
# 统计每一列数据类型
NA.Var
```

```python
# 这个代码可查看某个特征的数据类型
df['AASN'].dtypes
```

```python
# 写一个循环,把所有缺失列的数据类型添加到列表中
temp=[]
for i in NA.Var:
    temp.append(df[i].dtypes)
```

```python
temp
```

```python
NA['数据类型']=temp
```

```python
NA
```

缺失值填充策略：

分类(不管数值型还是字符串型)变量：通常分类水平出现的次数多，出现的概率就是最高的。用众数来填充。

连续性数值变量：age	c210b200	c210cip	c210hmi	c210hva	c210mah	c210psu	c210wht	ilor	meda 用均值填充

```python
# 单独取出缺失列观察下数据类型
df[list(NA.Var)].head(10)
```
