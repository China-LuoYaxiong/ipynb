## 新增列 平均薪资, 放在薪资的后面  
平均薪资为最高薪资和最低薪资的平均值, 比如10k-20k, 平均薪资是 **15000**

```python
df.head()
```

```python
a = df.薪资.str.split('-', expand=True)
a.columns = ['最低工资', '最高工资']
a
```

```python
def change(s):
    if 'k' in s:
        return float(s.replace('k', ''))*1000
    elif 'K' in s:
        return float(s.replace('K', ''))*1000
    else:
        pass
```

```python
a.最低工资.map(change).value_counts()
```

```python
a.最高工资.map(change).value_counts()
```

```python
a
```

```python
a['最低工资'] = a.最低工资.map(change)
a['最高工资'] = a.最高工资.map(change)
a['平均工资'] = (a.最低工资+a.最高工资)*0.5
```

```python
a
```

```python
df.insert(8, '平均薪资', a.平均工资)
```

## 将上诉平均薪资的分析过程封装成一个函数  
- 根据传入的列名的不同而进行分组  
- 计算该列名下不同类别的平均薪资的均值， 并按此进行降序输出, 如果类别太多， 只保留平均薪资的均值最高的前8个
- 利用直方图画出上述结果

```python
gp = df.groupby('城市')
a = gp['平均薪资'].agg(np.mean).sort_values(ascending = False)
a
```

```python
a.values
```

```python
a[0:8]
```

```python
def x(c):
    gp = df.groupby(c)
    a = gp['平均薪资'].agg(np.mean).sort_values(ascending = False)
    if len(a)<=8:
        plt.figure(figsize=(10, 6), dpi=100)
        plt.bar(range(len(a)), a.values)
        plt.xticks(range(len(a)), a.index, rotation=30)
        plt.title("各{}与平均薪资之间的关系".format(c))
    else:
        a = a[0:8]
        plt.figure(figsize=(10, 6), dpi=100)
        plt.bar(range(len(a)), a.values)
        plt.xticks(range(len(a)), a.index, rotation=30)
        plt.title("各{}与平均薪资之间的关系".format(c))
```

```python
# 导入地理类
from pyecharts.charts import Geo
# 导入配置项类
from pyecharts import options as opts
```

```python
geo = Geo()
```

```python
data = [('北京' , 250),
        ('上海',65),
        ('哈尔滨', 128),
        ('浙江', 25),
        ("杭州", 64), 
        ('大邑县',  100)]
```
