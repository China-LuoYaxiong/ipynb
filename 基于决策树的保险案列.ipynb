## 缺失值处理

### 首先统计每一列的数据类型

```python
NA  # 所有缺失的特征，共20个
```

```python
# 统计每一列数据类型
NA.Var
```

```python
# 这个代码可查看某个特征的数据类型
df['AASN'].dtypes
```

```python
# 写一个循环,把所有缺失列的数据类型添加到列表中
temp=[]
for i in NA.Var:
    temp.append(df[i].dtypes)
```

```python
temp
```

```python
NA['数据类型']=temp
```

```python
NA
```

缺失值填充策略：

分类(不管数值型还是字符串型)变量：通常分类水平出现的次数多，出现的概率就是最高的。用众数来填充。

连续性数值变量：age	c210b200	c210cip	c210hmi	c210hva	c210mah	c210psu	c210wht	ilor	meda 用均值填充

```python
# 单独取出缺失列观察下数据类型
df[list(NA.Var)].head(10)
```

### 去掉连续数值型的那些特征

```python
s = "age c210b200 c210cip c210hmi c210hva c210mah c210psu c210wht ilor meda"
list1 = s.split()
list1      # 这十个特征是连续数值的特征用均值填充
```

```python
L_fenlei = []
for i in list(NA.Var):
    if i not in list1:
        L_fenlei.append(i)
```

```python
L_fenlei  # 这十个特征是分类型特征，用众数填充
```

#### 对分类型特征进行众数填充

```python
for i in L_fenlei:
    df[i].fillna(df[i].mode()[0],inplace=True)
```

#### 对连续数值型特征进行均值填充

```python
for i in list1:
    df[i].fillna(df[i].mean(),inplace=True)
```

#### 最后检查一下是否还有缺失值

```python
df.isnull().sum()/df.shape[0]   #ok，结果显示没有了缺失值
```

### 将二分类以上的特征进one-hot编码

```python
# 统计一下有多少2分类，多少2分类以上的
two = 0
L_two_feature = []
L_over_two_feature = []
over_two =0
for i in df:
    if df[i].dtypes=='object':
        if df_copy[i].nunique()==2:
            two+=1
            L_two_feature.append(i)
        else:
            over_two+=1
            L_over_two_feature.append(i)
            
            
print(two)
print(over_two)
print(L_two_feature)   # 二分类的特征
print( L_over_two_feature)  # 超过两个分类 的特征
```

```python
df['GEND'].unique()  # 比如GEND就是二分类
```

```python
# map enumerate   
df['GEND'].map(dict([(j,i) for i,j in enumerate(df['GEND'].unique())]))  二分类就变成 0 1 
```

### 将所有二分类转变成 0 1

```python
for i in L_two_feature:
    df_copy[i]=df_copy[i].map(dict([(j,i) for i,j in enumerate(df_copy[i].unique())]))
```

```python
df_copy.head()
```

### 将超过二分类的进行one-hot编码

```python
L_over_two_feature
```

```python
# 转换为one-hot编码
dummies = pd.get_dummies(df[L_over_two_feature],drop_first=True)
```

```python
dummies.shape
```

```python
df_copy.shape
```

```python
df2 = df_copy.drop(columns=L_over_two_feature)
```

```python
df2.shape
```

```python
df_final = pd.concat([df2,dummies],axis=1,sort=False)
df_final.shape
```
