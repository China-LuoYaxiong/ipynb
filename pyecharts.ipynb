# 柱状图

```python
df = pd.DataFrame(data = [['Category 1', 50, 50, 200, 300,-300],
                         ['Category 2', 100, 110, 250, 460, -460],
                         ['Category 3', 200, 90, 200, 490, -490],
                         ['Category 4', 350, 50, 60, 460, -460],
                         ['Category 5', 50, 150, 400, 600, -600],
                         ['Category 6', 250, 50, 260, 560, -560]],
                columns = ['Category', 'Series1', 'Series2', 'Series3', 'Series4','Series5']
                )
df['合计'] = df.iloc[:, 1:].sum(axis=1)
df
```

```python
x = df.Category.to_list()
y1 = df.Series1.to_list()
y2 = df.Series2.to_list()
y3 = df.Series3.to_list()
y4 = df.Series4.to_list()
y5 = df.Series5.to_list()
y6 = df.合计.to_list()


from pyecharts.charts import Bar
# 类的实例化,实例化一个柱状图的对象
bar = Bar(
    # 初始化配置项
    init_opts = opts.InitOpts(
        page_title='01柱状图'  # 网页标题
        # ,width = '1100px'
        # ,height = '700px'
    )
)
# 添加横轴和纵轴的数据
bar.add_xaxis(xaxis_data=x)

bar.add_yaxis(series_name='Series1'
              ,y_axis=y1 
              ,color='#028599' # 绿色
             )   

bar.add_yaxis(series_name='Series2'
              ,y_axis=y2 
              ,color='#FF3F5F' # 红色
             )  

bar.add_yaxis(series_name='Series3'
              ,y_axis=y3
              ,color='#FEB834'  # 黄色
             )  

# 设置图表全局配置项
bar.set_global_opts(
    # 图表标题
    title_opts=opts.TitleOpts(is_show=True, 
                              title='这是一个柱状图', 
                              pos_left='center'
                             )
    # 图例
    ,legend_opts=opts.LegendOpts(is_show=True, 
                                pos_top=30,  # 图例离上边界 30
                                border_color='transparent'# 图例边界颜色为 透明
                               
                               )
    
    # X 坐标轴配置项
    ,xaxis_opts=opts.AxisOpts(type_='category'     # 'value': 数值轴，适用于连续数据。
                                                  # 'category': 类目轴，适用于离散的类目数据，为该类型时必须通过 data 设置类目数据。
                                                  # 'time': 时间轴，适用于连续的时序数据，与数值轴相比时间轴带有时间的格式化，在刻度计算上也有所不同，
                                                  # 例如会根据跨度的范围来决定使用月，星期，日还是小时范围的刻度。
                                                  # 'log' 对数轴。适用于对数数据。
                             ,name = 'X坐标轴名称' # 轴名称
                             ,is_show = True      # 是否显示x轴
                             ,name_location='center'  # 'start', 'middle' 或者 'center','end'
                             ,name_gap = 35      # 坐标轴名称与轴线之间的距离。
                             # 轴名称文字样式配置
                             ,name_textstyle_opts=opts.TextStyleOpts(# color='black'       # 文字颜色，使用默认
                                                                     # font_style='normal' # 字体风格
                                                                     # ,font_weight='normal'   # 字体粗细
                                                                     # ,font_family='Microsoft YaHei'        # 字体 'serif' , 'monospace', 'Arial', 'Courier New', 'Microsoft YaHei',
                                                                     font_size=15  # 字体大小
                                                                    
                                                                    
                                                                    )
                            
                            
                            
                            )
    #  Y 坐标轴配置项
    ,yaxis_opts=opts.AxisOpts(type_='value'     # 'value': 数值轴，适用于连续数据。
                                                  # 'category': 类目轴，适用于离散的类目数据，为该类型时必须通过 data 设置类目数据。
                                                  # 'time': 时间轴，适用于连续的时序数据，与数值轴相比时间轴带有时间的格式化，在刻度计算上也有所不同，
                                                  # 例如会根据跨度的范围来决定使用月，星期，日还是小时范围的刻度。
                                                  # 'log' 对数轴。适用于对数数据。
                             ,name = 'Y坐标轴名称' # 轴名称
                             ,is_show = True      # 是否显示
                             ,name_location='center'  # 'start', 'middle' 或者 'center','end'
                             ,name_gap = 35      # 坐标轴名称与轴线之间的距离。
                             # 轴名称文字样式配置
                             ,name_textstyle_opts=opts.TextStyleOpts(# color='black'       # 文字颜色,使用默认
                                                                     # font_style='normal' # 字体风格
                                                                     # font_weight='normal'   # 字体粗细
                                                                     # ,font_family='Microsoft YaHei'        # 字体 'serif' , 'monospace', 'Arial', 'Courier New', 'Microsoft YaHei',
                                                                     font_size=15  # 字体大小
                                                                    
                                                                    
                                                                    )
                             )
    
                   
)

# 设置系列配置项
bar.set_series_opts(
    # 数据标签配置项
    label_opts=opts.LabelOpts(
        is_show = True,
        position = 'top'
    )
    
    # 标记线配置项
    ,markline_opts=opts.MarkLineOpts(
        data=[
                 opts.MarkLineItem(y=240, 
                                   name="yAxis=240", 
                                   # 线样式配置项
                                   linestyle_opts = opts.LineStyleOpts(color='#5470C6' # 线的颜色
                                                                       ,type_='dashed'  # 线型
                                                                      
                                                                      )
                                  
                                  )
            
            
                ,opts.MarkLineItem(y=150, name="yAxis=150")
             
             
             ]
    )
)

# 在jupyter 中显示
bar.load_javascript()
# jupyter渲染的图片往往不够清晰，导出再截图
html_name = '01柱状图.html'
bar.render(html_name)
```

# 对比条形图

```python
from pyecharts.charts import Bar
from pyecharts import options as opts
from pyecharts.commons.utils import JsCode

# 准备数据
x_data = ['A1', 'B1', 'C1', 'D1', 'E1', 'F1']
y_data1 = [20, 30, 40, 50, 60, 100]
y_data2 = [15, 25, 35, 45, 55, 100]

# 创建Bar实例
bar = Bar(
    # 初始化配置项
    init_opts = opts.InitOpts(
        page_title='06对比条形图'  # 网页标题
        # ,width = '1100px'
        # ,height = '700px'
    )

)

# 添加数据
bar.add_xaxis(x_data)
bar.add_yaxis("Series1", 
              y_data1, 
              stack="stack1", 
              color='#028599', 
              label_opts=opts.LabelOpts(formatter="{c}")) # c代表限制的是 值


bar.add_yaxis("Series2", 
              [-i for i in y_data2], 
              color='#FF3F5F', 
              stack="stack1",  
              label_opts=opts.LabelOpts(is_show=True,
                                           # 利用回调函数去除负号
                                           formatter=JsCode(
                                               """
                                               function(params) {
                                                   return -(params.data)
                                               }
                                               """

                                           )


                                          ))
bar.reversal_axis()

# 设置全局配置项
bar.set_global_opts(
    # 图表标题
    title_opts=opts.TitleOpts(is_show=True, 
                              title='这是一个对比条形图', 
                              pos_left='center'
                             ),
    # 图例
    legend_opts=opts.LegendOpts(is_show=True, 
                                pos_top=30,  # 图例离上边界 30
                                border_color='transparent'# 图例边界颜色为 透明
                               
                               ),  
    
    
    
    
    # x坐标轴不显示
    xaxis_opts=opts.AxisOpts(is_show=False),
    
    # y轴
    yaxis_opts=opts.AxisOpts(is_show=True, 
                            # 坐标轴刻度线配置项
                            axisline_opts = opts.AxisLineOpts(is_show=False),
                             
                            # 坐标轴刻度配置项
                            axistick_opts = opts.AxisTickOpts(is_show=False),
                             
                            # 分割线配置项
                            splitline_opts=opts.SplitLineOpts(is_show=True)
                            )
)

# 渲染图表
bar.load_javascript()
html_name = '05对比条形图.html'
bar.render(html_name)

# render_notebook()显示不出来，就用下面这种# bar.render_notebook()
from IPython.display import IFrame
IFrame(src=html_name, width=950, height=520)
```

## 散点图和线图

```python
from pyecharts import options as opts
from pyecharts.charts import Line, Scatter
from pyecharts.faker import Faker

np.random.seed(0)
x = np.random.randn(50)
y1 = x+np.random.rand(50)

scatter = (
    Scatter()
    .add_xaxis(x)
    .add_yaxis("", y1)
)



X = pd.DataFrame(x).values
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y1)
y2 = lr.coef_[0]*x+lr.intercept_
line = (
    Line()
    .add_xaxis(x)
    .add_yaxis("", y2,is_smooth=True, is_symbol_show=False, color='#028599',
               linestyle_opts = opts.LineStyleOpts(width=3,  color='#FF3F5F',)
              
              )
    .set_global_opts(title_opts=opts.TitleOpts(title="Overlap-line+scatter"))
)

scatter.overlap(line)


# 全局配置项
scatter.set_global_opts(
    
    # 图表标题
    title_opts=opts.TitleOpts(is_show=True, 
                              title='散点图+线图', 
                              pos_left='center'
                             ),
    # 图例
    legend_opts=opts.LegendOpts(is_show=False 
                                ,pos_top=30  # 图例离上边界 30
                                # ,pos_left=50
                                # ,orient = 'vertical'   #  可选：'horizontal', 'vertical'#
                                ,border_color='transparent'# 图例边界颜色为 透明
                               
                               ),  
    
    # x轴配置项
    xaxis_opts=opts.AxisOpts(
        # x轴数值为数值型
        type_="value",     #  如果x是数值型，这里要指定为 value
    )
)

# 系列配置项
scatter.set_series_opts(
    # 标签配置项
    label_opts=opts.LabelOpts(
        is_show=False
    )
)


scatter.render("散点图和线图.html")
from IPython.display import IFrame
IFrame(src="散点图和线图.html", width=950, height=520)
```
